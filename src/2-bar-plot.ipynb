### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']  # Different colors for each bar

# Error handling: Check if all lists have the same length
if len(categories) != len(values) or len(categories) != len(colors):
    raise ValueError("Categories, values, and colors lists must have the same length.")

# Error handling: Check if values contain only numbers
if not all(isinstance(v, (int, float)) for v in values):
    raise ValueError("Values list must contain only numeric data.")

# Create the bar plot
plt.bar(categories, values, color=colors)

# Customize the plot
plt.title("Bar Plot of Categories vs Values")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

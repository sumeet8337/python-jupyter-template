### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

try:
    # Generate x values from 0 to 10
    x = np.linspace(0, 10, 100)
    y = x ** 2  # Square of x
except Exception as e:
    print(f"Error in data generation: {e}")
    x, y = [], []  # Default empty lists in case of failure

try:
    # Create the plot
    plt.plot(x, y, label="y = x²", color='b')

    # Customize the plot
    plt.title("Line Plot of y = x²")
    plt.xlabel("X values")
    plt.ylabel("Y values (Squared)")
    plt.grid(True)
    plt.legend()

    # Show the plot
    plt.show()
except Exception as e:
    print(f"Error during plot creation: {e}")
```

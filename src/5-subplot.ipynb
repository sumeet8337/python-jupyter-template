### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plots
try:
    x = np.linspace(0, 10, 100)
    y = x ** 2  # Line chart data

    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]  # Bar chart data

    scatter_x = [1, 2, 3, 4, 5]
    scatter_y = [2, 4, 6, 8, 10]  # Scatter plot data

    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]  # Pie chart data

    # Validation checks
    if len(categories) != len(values):
        raise ValueError("Bar chart data mismatch: categories and values must have the same length.")
    
    if len(scatter_x) != len(scatter_y):
        raise ValueError("Scatter plot data mismatch: x and y values must have the same length.")
    
    if len(labels) != len(sizes):
        raise ValueError("Pie chart data mismatch: labels and sizes must have the same length.")
    
    if sum(sizes) <= 0:
        raise ValueError("Pie chart sizes must have a total greater than zero.")

    # Create a 2x2 subplot layout
    fig, axes = plt.subplots(2, 2, figsize=(10, 10))

    # First subplot - Line chart
    axes[0, 0].plot(x, y, color='b', label='y = x²')
    axes[0, 0].set_title("Line Chart")
    axes[0, 0].set_xlabel("X values")
    axes[0, 0].set_ylabel("Y values")
    axes[0, 0].legend()
    axes[0, 0].grid(True)

    # Second subplot - Bar chart
    axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
    axes[0, 1].set_title("Bar Chart")
    axes[0, 1].set_xlabel("Categories")
    axes[0, 1].set_ylabel("Values")
    axes[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

    # Third subplot - Scatter plot
    axes[1, 0].scatter(scatter_x, scatter_y, color='blue')
    axes[1, 0].set_title("Scatter Plot")
    axes[1, 0].set_xlabel("X values")
    axes[1, 0].set_ylabel("Y values")
    axes[1, 0].grid(True)

    # Fourth subplot - Pie chart
    axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'], startangle=140)
    axes[1, 1].set_title("Pie Chart")

    # Adjust layout for better spacing
    plt.tight_layout()

    # Show the plot
    plt.show()

except ValueError as e:
    print(f"Error: {e}")

except Exception as e:
    print(f"Unexpected error: {e}")
```
